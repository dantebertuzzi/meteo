```julia
using CSV, DataFrames, Dates, PlotlyJS
```

```julia
# Carregar os dados do arquivo CSV (substitua pelo caminho correto do arquivo)
clima_data = CSV.read("INMET_NE_PE_A307_PETROLINA_01-01-2013_A_31-12-2013.CSV", DataFrame; delim=';')
```

```julia
names(clima_data)
```

```julia
# Renomear as colunas de acordo com os novos nomes
rename!(clima_data, Dict(
    "DATA (YYYY-MM-DD)" => :DATA,
    "HORA (UTC)" => :HORA_UTC,
    "PRECIPITA\xc7\xc3O TOTAL, HOR\xc1RIO (mm)" => :PRECIPITACAO_TOTAL_HORARIO_mm,
    "PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)" => :PRESSAO_ATMOSFERICA_NIVEL_ESTACAO_HORARIA_mB,
    "PRESS\xc3O ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)" => :PRESSAO_ATMOSFERICA_MAX_HORA_ANT_AUT_mB,
    "PRESS\xc3O ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)" => :PRESSAO_ATMOSFERICA_MIN_HORA_ANT_AUT_mB,
    "RADIACAO GLOBAL (KJ/m\xb2)" => :RADIACAO_GLOBAL_KJ_m2,
    "TEMPERATURA DO AR - BULBO SECO, HORARIA (\xb0C)" => :TEMPERATURA_DO_AR_BULBO_SECO_HORARIA_C,
    "TEMPERATURA DO PONTO DE ORVALHO (\xb0C)" => :TEMPERATURA_PONTO_ORVALHO_C,
    "TEMPERATURA M\xc1XIMA NA HORA ANT. (AUT) (\xb0C)" => :TEMPERATURA_MAXIMA_HORA_ANT_AUT_C,
    "TEMPERATURA M\xcdNIMA NA HORA ANT. (AUT) (\xb0C)" => :TEMPERATURA_MINIMA_HORA_ANT_AUT_C,
    "TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (\xb0C)" => :TEMPERATURA_ORVALHO_MAX_HORA_ANT_AUT_C,
    "TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (\xb0C)" => :TEMPERATURA_ORVALHO_MIN_HORA_ANT_AUT_C,
    "UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)" => :UMIDADE_REL_MAX_HORA_ANT_AUT,
    "UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)" => :UMIDADE_REL_MIN_HORA_ANT_AUT,
    "UMIDADE RELATIVA DO AR, HORARIA (%)" => :UMIDADE_RELATIVA_DO_AR_HORARIA,
    "VENTO, DIRE\xc7\xc3O HORARIA (gr) (\xb0 (gr))" => :VENTO_DIRECAO_HORARIA_gr,
    "VENTO, RAJADA MAXIMA (m/s)" => :VENTO_RAJADA_MAXIMA_m_s,
    "VENTO, VELOCIDADE HORARIA (m/s)" => :VENTO_VELOCIDADE_HORARIA_m_s
))
```

```julia
names(clima_data)
```

```julia
# Convertendo a coluna DATA para o tipo Date
clima_data.DATA = Date.(clima_data.DATA, "dd/mm/yyyy")
```

```julia
# Convertendo a coluna de temperatura para numérico e substituindo vírgulas por pontos
clima_data.TEMPERATURA_DO_AR_BULBO_SECO_HORARIA_C = parse.(Float64, replace.(clima_data.TEMPERATURA_DO_AR_BULBO_SECO_HORARIA_C, "," => "."))
```

```julia
using PlotlyJS
```

```julia
# Criando o gráfico com PlotlyJS
scatter_plot = PlotlyJS.Plot(
    PlotlyJS.scatter(
        x=clima_data.DATA,
        y=clima_data.TEMPERATURA_DO_AR_BULBO_SECO_HORARIA_C,
        mode="markers",
        marker_color="blue",
        opacity=0.6
    ),
    Layout(
        title="Temperatura Horária (Bulbo Seco)",
        xaxis_title="Data",
        yaxis_title="Temperatura [°C]",
        yaxis=attr(
            range=[0, 50],
            side="left",
            titlefont=attr(color="blue")
        ),
        yaxis2=attr(
            title="Temperatura [°F]",
            overlaying="y",
            side="right",
            tickvals=[32, 50, 68, 86, 104, 122],
            ticktext=["32", "50", "68", "86", "104", "122"]
        ),
        template="plotly_white"
    )
)

# Mostrar o gráfico
display(scatter_plot)
```
