> ### Questão 1 `probabilidade binomial`
Em uma determinada cidade, a probabilidade de um morador, escolhido ao acaso, ter seguro de carro é de 80%. Em um determinado dia, cinco moradores são escolhidos ao acaso.

A probabilidade de que exatamente **3** deles tenham seguro de carro é, aproximadamente?

### Solução:

Esse é um problema de `probabilidade binomial`, onde temos uma **série de experimentos independentes** (no caso, selecionar cinco moradores) e **dois resultados possíveis para cada experimento** (ter ou não ter seguro de carro).

Os parâmetros da distribuição binomial são:

- $n = 5$ (número de experimentos, ou seja, cinco moradores),
- $p = 0,8$ (probabilidade de sucesso, ou seja, a probabilidade de um morador ter seguro de carro),
- $k = 3$ (número de sucessos desejados, ou seja, queremos exatamente três moradores com seguro de carro).

A fórmula da probabilidade binomial é dada por:

$$
P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Onde:

$$
\binom{n}{k} = \frac{n!}{k!(n - k)!}
$$

Substituindo os valores $n = 5$, $p = 0,8$, e $k = 3$, temos:

$$
P(X = 3) = \binom{5}{3} \cdot (0,8)^3 \cdot (0,2)^2
$$

Calculando passo a passo:

$$
\binom{5}{3} = \frac{5!}{3!(5 - 3)!} = \frac{5 \times 4}{2 \times 1} = 10
$$

$$
(0,8)^3 = 0,512 \quad \text{e} \quad (0,2)^2 = 0,04
$$

Agora multiplicamos os resultados:

$$
P(X = 3) = 10 \times 0,512 \times 0,04 = 0,2048
$$

Portanto, a probabilidade de que exatamente 3 dos 5 moradores tenham seguro de carro é aproximadamente **0,2048** ou **20,48%**.


> ### Questão 2

Uma empresa resolveu analisar a percepção que os seus 30 principais fornecedores tinham da empresa. Para isso, foi enviado um questionário com algumas perguntas. Uma dessas perguntas solicitava que os fornecedores atribuíssem uma nota de 0 a 10 para o prazo de pagamento aos fornecedores. Após os 30 fornecedores responderem, obteve-se uma nota **média de 7,5**. Retirando-se as **cinco maiores notas**, a média das 25 notas restantes cai para **7,0**.

Qual é a média dessas cinco maiores notas atribuídas pelos fornecedores?

### Solução:

Vamos resolver o problema passo a passo, utilizando as informações fornecidas:

#### Primeira média (com as 30 notas):

- O total de fornecedores é 30.
- A média dessas 30 notas é 7,5.
  
Isso significa que a soma das 30 notas é:

$$
S_{30} = 30 \times 7,5 = 225
$$

#### Segunda média (com as 25 notas restantes):

- Ao retirar as 5 maiores notas, a média das 25 notas restantes cai para 7,0.
  
Portanto, a soma das 25 notas restantes é:

$$
S_{25} = 25 \times 7,0 = 175
$$

#### Soma das 5 maiores notas:

A soma das 5 maiores notas será a diferença entre a soma total das 30 notas e a soma das 25 notas restantes:

$$
S_{5} = S_{30} - S_{25} = 225 - 175 = 50
$$

#### Média das 5 maiores notas:

A média dessas 5 maiores notas é a soma delas dividida por 5:

$$
\text{Média das 5 maiores notas} = \frac{S_{5}}{5} = \frac{50}{5} = 10
$$

Portanto, a média das cinco maiores notas atribuídas pelos fornecedores é **10**.


> ### Questão 3

Na fila de um cartório estão **10 pessoas**: uma delas possui **17 anos**, duas possuem **19 anos**, quatro possuem **20 anos**, uma possui **21 anos** e duas possuem **22 anos**. Escolhendo-se, aleatoriamente, uma pessoa dessa fila, a probabilidade de que sua idade seja maior do que a mediana do grupo será?

### Solução:

Vamos resolver o problema passo a passo:

#### Organizando as idades em ordem crescente:

Temos as idades das 10 pessoas: 17, 19, 19, 20, 20, 20, 20, 21, 22, 22.  
As idades estão organizadas em ordem crescente:  
$$ 17, 19, 19, 20, 20, 20, 20, 21, 22, 22 $$

#### Calculando a mediana:

A mediana é o valor central de um conjunto de dados ordenados. Para um número par de dados, a mediana é a média dos dois valores centrais.  
No caso de 10 pessoas, temos 10 valores. Os dois valores centrais são o 5º e o 6º:  
- O 5º valor é 20.  
- O 6º valor também é 20.  

Portanto, a mediana é:

$$
\text{Mediana} = \frac{20 + 20}{2} = 20
$$

#### Identificando as idades maiores que a mediana:

As idades maiores que 20 são: 21, 22, 22.  
Portanto, há 3 pessoas com idades maiores que a mediana.

#### Calculando a probabilidade:

A probabilidade é a razão entre o número de pessoas com idades maiores que a mediana e o número total de pessoas.  
O número total de pessoas é 10, e o número de pessoas com idades maiores que a mediana é 3.

A probabilidade será:

$$
P(\text{idade} > \text{mediana}) = \frac{3}{10} = 0,3
$$

Portanto, a probabilidade de escolher uma pessoa cuja idade seja maior do que a mediana do grupo é **0,3** ou **30%**.


> ### Questão 4 `Teorema de Bayes`

`Danilo` diariamente come uma coxinha em frente à faculdade. A coxinha é assada de forma aleatória por um dos três cozinheiros da lanchonete: 40% das vezes, a coxinha é assada por `Roberta`; 40% das vezes, por `Pedro`; e 20% das vezes, por `Carol`. `Roberta` assa mal o salgado e acaba deixando o centro da coxinha frio em 10% das vezes; `Pedro` o faz em 5% das vezes e `Carol` 20% das vezes. Certo dia, `Danilo` pede sua coxinha e, ao experimentá-la, verifica que o centro está frio.

A probabilidade de que essa coxinha tenha sido feita por `Pedro` é igual a:

### Solução:

Essa é uma questão de `probabilidade condicional`, e podemos resolvê-la utilizando o `Teorema de Bayes`. Vamos definir os eventos:

- $C_P$: A coxinha foi assada por Pedro.
- $C_R$: A coxinha foi assada por Roberta.
- $C_C$: A coxinha foi assada por Carol.
- $F$: O centro da coxinha está frio.

Queremos calcular $P(C_P \mid F)$, ou seja, a probabilidade de que Pedro tenha assado a coxinha, dado que o centro está frio.

O Teorema de Bayes é dado por:

$$
P(C_P \mid F) = \frac{P(F \mid C_P) \cdot P(C_P)}{P(F)}
$$

Agora, vamos calcular os componentes dessa fórmula:

#### Probabilidade de Pedro ter assado a coxinha $P(C_P)$:
$$
P(C_P) = 0,40 \quad
$$

#### Probabilidade de o centro estar frio dado que Pedro assou a coxinha $P(F \mid C_P)$:
Pedro deixa o centro frio em 5% das vezes, então:
$$
P(F \mid C_P) = 0,05.
$$

#### Probabilidade total de o centro estar frio $P(F)$:

A probabilidade total de o centro estar frio é obtida somando as probabilidades de o centro estar frio, ponderadas pelas probabilidades de cada cozinheiro ter assado a coxinha:
$$
P(F) = P(F \mid C_R) \cdot P(C_R) + P(F \mid C_P) \cdot P(C_P) + P(F \mid C_C) \cdot P(C_C)
$$

Substituindo os valores:

- $P(F \mid C_R) = 0,10$ (Roberta deixa o centro frio em 10% das vezes),
- $P(C_R) = 0,40$ (40% das vezes a coxinha é assada por Roberta),
- $P(F \mid C_P) = 0,05$ (Pedro deixa o centro frio em 5% das vezes),
- $P(C_P) = 0,40$ (40% das vezes a coxinha é assada por Pedro),
- $P(F \mid C_C) = 0,20$ (Carol deixa o centro frio em 20% das vezes),
- $P(C_C) = 0,20$ (20% das vezes a coxinha é assada por Carol).

Agora, calculando $P(F)$:

$$
P(F) = (0,10 \times 0,40) + (0,05 \times 0,40) + (0,20 \times 0,20) = 0,04 + 0,02 + 0,04 = 0,10
$$

#### Aplicando o Teorema de Bayes:

Agora podemos calcular $P(C_P \mid F)$:

$$
P(C_P \mid F) = \frac{P(F \mid C_P) \cdot P(C_P)}{P(F)} = \frac{0,05 \times 0,40}{0,10} = \frac{0,02}{0,10} = 0,20
$$

Portanto, a probabilidade de que a coxinha tenha sido feita por Pedro, dado que o centro está frio, é **0,20** ou **20%**.



> ### Questão 5

Em um encontro regional de motoqueiros estavam presentes 20 motoqueiros, cada um com sua moto. A tabela abaixo apresenta a faixa etária (em anos) e a quantidade de motos que se enquadram naquela faixa. De acordo com os dados da tabela, podemos determinar que a idade média das motos presentes é igual a:

| Idade | Número de veículos |
|-------|---------------------|
| 1 -- 3  | 7                   |
| 3 -- 5  | 5                   |
| 5 -- 7  | 4                   |
| 7 -- 9  | 3                   |
| 9 -- 11 | 1                   |

### Solução:

Vamos calcular a idade média das motos presentes utilizando os dados da tabela. Para isso, seguimos os seguintes passos:

#### 1. Definir as faixas etárias e o número de veículos por faixa:

- Faixa etária (anos) | Número de veículos
  - 1 -- 3 | 7
  - 3 -- 5 | 5
  - 5 -- 7 | 4
  - 7 -- 9 | 3
  - 9 -- 11 | 1

#### 2. Encontrar o ponto médio de cada faixa etária:

O ponto médio de uma faixa é a média dos limites inferior e superior da faixa etária. Vamos calcular:

- Para a faixa 1 -- 3, o ponto médio é:
  $$
  \frac{1 + 3}{2} = 2
  $$

- Para a faixa 3 -- 5, o ponto médio é:
  $$
  \frac{3 + 5}{2} = 4
  $$

- Para a faixa 5 -- 7, o ponto médio é:
  $$
  \frac{5 + 7}{2} = 6
  $$

- Para a faixa 7 -- 9, o ponto médio é:
  $$
  \frac{7 + 9}{2} = 8
  $$

- Para a faixa 9 -- 11, o ponto médio é:
  $$
  \frac{9 + 11}{2} = 10
  $$

#### 3. Multiplicar o ponto médio pelo número de veículos em cada faixa etária:

Agora multiplicamos o ponto médio pela quantidade de veículos em cada faixa etária:

- $2 \times 7 = 14$
- $4 \times 5 = 20$
- $6 \times 4 = 24$
- $8 \times 3 = 24$
- $10 \times 1 = 10$

#### 4. Somar os resultados das multiplicações:

Somamos os produtos obtidos no passo anterior:

$$
14 + 20 + 24 + 24 + 10 = 92
$$

#### 5. Calcular a média:

Para calcular a idade média, dividimos a soma dos produtos pelo número total de veículos:

$$
\text{Idade média} = \frac{92}{20} = 4,6
$$

Portanto, a idade média das motos presentes no encontro é **4,6 anos**.


> ### Questão 6 `Probabilidade da União de dois Eventos`

Após uma escalada, a probabilidade de que a corda precise de reparos é de 0,40; a probabilidade de que algum mosquetão tenha de ser trocado é de 0,10; a probabilidade de a corda precisar de reparos e o mosquetão ter de ser trocado é igual a 0,05. Assim, a probabilidade de a corda precisar de reparos ou o mosquetão ter de ser trocado será igual a:

### Solução:

Essa questão envolve a probabilidade da união de dois eventos, que podemos calcular usando a seguinte fórmula:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

Onde:

- $P(A)$ é a probabilidade de a corda precisar de reparos.
- $P(B)$ é a probabilidade de o mosquetão precisar ser trocado.
- $P(A \cap B)$ é a probabilidade de ambos os eventos acontecerem, ou seja, a corda precisar de reparos e o mosquetão precisar ser trocado.

De acordo com o enunciado:

- $P(A) = 0,40$ (probabilidade de a corda precisar de reparos),
- $P(B) = 0,10$ (probabilidade de o mosquetão precisar ser trocado),
- $P(A \cap B) = 0,05$ (probabilidade de ambos ocorrerem).

Agora, substituímos os valores na fórmula:

$$
P(A \cup B) = 0,40 + 0,10 - 0,05
$$

$$
P(A \cup B) = 0,50 - 0,05
$$

$$
P(A \cup B) = 0,45
$$

Portanto, a probabilidade de que a corda precise de reparos ou o mosquetão tenha de ser trocado é **0,45**, ou **45%**.

- Para $y$: Temos $y^2 - 4y$. Para completar o quadrado, tomamos $\left(\frac{-4}{2}\right)^2 = 4$. Então, adicionamos e subtraímos 4:
